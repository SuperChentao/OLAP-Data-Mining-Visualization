```python
# process data in fact table, the joined table attribute will be process in another file
import pandas as pd
```

```python
input_data_path='./DataMart/'
otout_data_path='./NewDataMart/'

loan_table=pd.read_csv(input_data_path+"loan_fact_table.csv")
loan_table
```

```python

# saving account attribute, purpose attribute, checking account attribute are ordinal
# one hot encoding to saving account attribute
saving_account_dimension=pd.read_csv(input_data_path+"saving_account_dimension.csv")
print("before one hot encoding:")
display(saving_account_dimension)

saving_account_dimension['savings']=pd.Categorical(saving_account_dimension['savings'],categories=['unknown_no_saving_acc', 'below_100','below_1000','above_1000','below_500'], ordered=True)
one_hot_encoding=pd.get_dummies(saving_account_dimension['savings']).astype(int)
new_saving_account_dimension=pd.concat([saving_account_dimension,one_hot_encoding], axis=1)
new_saving_account_dimension.drop(columns=['savings'], inplace=True)
print("after one hot encoding")
display(new_saving_account_dimension)
new_saving_account_dimension.to_csv(otout_data_path+"saving_account_dimension.csv", index=False)
```

```python
# one hot encoding to checking account attribute
checking_account_dimension=pd.read_csv(input_data_path+"checking_account_dimension.csv")
print("Before one hot encoding")
display(checking_account_dimension)

checking_account_dimension['status']=pd.Categorical(checking_account_dimension['status'],['below_0','below_200','no_cheking_acc','above'],ordered=True)
one_hot_encode=pd.get_dummies(checking_account_dimension['status']).astype(int)
new_checking_account_dimension=pd.concat([checking_account_dimension,one_hot_encode],axis=1)
new_checking_account_dimension.drop(columns=['status'],inplace=True)
print("After one hot encoding")
display(new_checking_account_dimension)
new_checking_account_dimension.to_csv(otout_data_path+"checking_account_dimension.csv", index=False)
```

```python
# one hot encoding to purpose attribute
attribute_dimension=pd.read_csv(input_data_path+"purpose_dimension.csv")
print("Before one hot encoding:")
display(attribute_dimension)

attribute_dimension['purpose']=pd.Categorical(attribute_dimension['purpose'],['radio_tv','education','furniture_equipment','car_new','car_used','business','domestic_applience'
,'repairs','retraining','others'], ordered=True)
one_hot_encode=pd.get_dummies(attribute_dimension['purpose']).astype(int)
new_purpose_dimension=pd.concat([attribute_dimension,one_hot_encode],axis=1)
new_purpose_dimension.drop(columns=['purpose'],inplace=True)
print("After one hot encoding")
display(new_purpose_dimension)
new_purpose_dimension.to_csv(otout_data_path+'new_purpose_dimension.csv')
```

```python
# Normalize the numeric attributes: loan_amt, duration
max_loan=loan_table['loan_amt'].max()
min_loan=loan_table['loan_amt'].min()
loan_table['normalized loan_amt']=loan_table['loan_amt'].apply(lambda x : ((x-min_loan)/(max_loan-min_loan)).round(3))

max_duration=loan_table['duration'].max()
min_duration=loan_table['duration'].min()
loan_table['normalized duration']=loan_table['duration'].apply(lambda x : ((x-min_duration)/(max_duration-min_duration)).round(3))
display(loan_table)
```
